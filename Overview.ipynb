```python
import matplotlib.pyplot as plt
import networkx as nx

# Cration du graphe dirig
G = nx.DiGraph()

# Dfinition des nuds avec des positions claires et espaces
nodes = {
    "Ontologie": (2, 5),
    "Base de Connaissance": (2, 4),
    "Moteur d'infrence": (4, 4),
    "Interface : Humain Machine": (4, 3),
    "Prdiction : Capteurs..": (6, 4),
    "Best Prdiction": (1, 3),
    "Bad Prdiction": (3, 3)
}

G.add_nodes_from(nodes.keys())

# Dfinition des connexions entre les nuds
edges = [
    ("Ontologie", "Base de Connaissance"),
    ("Ontologie", "Moteur d'infrence"),
    ("Base de Connaissance", "Best Prdiction"),
    ("Base de Connaissance", "Bad Prdiction"),
    ("Moteur d'infrence", "Prdiction : Capteurs.."),
    ("Moteur d'infrence", "Interface : Humain Machine"),
    ("Interface : Humain Machine", "Base de Connaissance"),
    ("Interface : Humain Machine", "Prdiction : Capteurs..")
]

G.add_edges_from(edges)

# Dfinition de la figure avec des espacements bien clairs
plt.figure(figsize=(8, 6))
pos = nodes  # Utilisation de positions fixes pour meilleure lisibilit

# Style des nuds et du texte
nx.draw_networkx_nodes(G, pos, node_color="lightblue", edgecolors="black", node_size=2500)
nx.draw_networkx_edges(G, pos, edge_color="black", arrows=True, arrowsize=15, width=1.5)
nx.draw_networkx_labels(G, pos, font_size=10, font_weight="bold", font_color="black")

# Suppression des axes pour un meilleur rendu
plt.axis("off")

# Sauvegarde sous format image
plt.savefig("diagram_ontology_simple.png", dpi=300, bbox_inches="tight")
plt.show()
```
