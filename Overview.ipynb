```python
import matplotlib.pyplot as plt
import networkx as nx

# Création du graphe dirigé
G = nx.DiGraph()

# Définition des nœuds avec des positions claires et espacées
nodes = {
    "Ontologie": (2, 5),
    "Base de Connaissance": (2, 4),
    "Moteur d'inférence": (4, 4),
    "Interface : Humain Machine": (4, 3),
    "Prédiction : Capteurs..": (6, 4),
    "Best Prédiction": (1, 3),
    "Bad Prédiction": (3, 3)
}

G.add_nodes_from(nodes.keys())

# Définition des connexions entre les nœuds
edges = [
    ("Ontologie", "Base de Connaissance"),
    ("Ontologie", "Moteur d'inférence"),
    ("Base de Connaissance", "Best Prédiction"),
    ("Base de Connaissance", "Bad Prédiction"),
    ("Moteur d'inférence", "Prédiction : Capteurs.."),
    ("Moteur d'inférence", "Interface : Humain Machine"),
    ("Interface : Humain Machine", "Base de Connaissance"),
    ("Interface : Humain Machine", "Prédiction : Capteurs..")
]

G.add_edges_from(edges)

# Définition de la figure avec des espacements bien clairs
plt.figure(figsize=(8, 6))
pos = nodes  # Utilisation de positions fixes pour meilleure lisibilité

# Style des nœuds et du texte
nx.draw_networkx_nodes(G, pos, node_color="lightblue", edgecolors="black", node_size=2500)
nx.draw_networkx_edges(G, pos, edge_color="black", arrows=True, arrowsize=15, width=1.5)
nx.draw_networkx_labels(G, pos, font_size=10, font_weight="bold", font_color="black")

# Suppression des axes pour un meilleur rendu
plt.axis("off")

# Sauvegarde sous format image
plt.savefig("diagram_ontology_simple.png", dpi=300, bbox_inches="tight")
plt.show()
```
