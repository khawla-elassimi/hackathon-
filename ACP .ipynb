```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import chi2_contingency

# Nettoyage des données (conversion en string pour éviter les erreurs)
df["Code d’anomalie"] = df["Code d’anomalie"].astype(str)
df["Gravité"] = df["Gravité"].astype(str)
df["Type"] = df["Type"].astype(str)  # "Diagnostic" ou "Event"

# Séparation des diagnostics et des événements
df_diag = df[df["Type"] == "Diagnostic"]
df_event = df[df["Type"] == "Event"]

# 1. Tableau croisé dynamique : Relation Gravité vs Type par Code d'Anomalie
pivot_diag = pd.crosstab(index=df_diag["Code d’anomalie"], columns=df_diag["Gravité"])
pivot_event = pd.crosstab(index=df_event["Code d’anomalie"], columns=df_event["Gravité"])

print("📊 Tableau croisé dynamique - Diagnostics :")
print(pivot_diag)
print("\n📊 Tableau croisé dynamique - Events :")
print(pivot_event)

# 2. Visualisation avec Heatmaps
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

sns.heatmap(pivot_diag, cmap="coolwarm", annot=True, fmt="d", linewidths=0.5, ax=axes[0])
axes[0].set_title("Relation Gravité vs Code d'Anomalie (Diagnostics)")
axes[0].set_xlabel("Gravité")
axes[0].set_ylabel("Code d'Anomalie")

sns.heatmap(pivot_event, cmap="coolwarm", annot=True, fmt="d", linewidths=0.5, ax=axes[1])
axes[1].set_title("Relation Gravité vs Code d'Anomalie (Events)")
axes[1].set_xlabel("Gravité")
axes[1].set_ylabel("Code d'Anomalie")

plt.tight_layout()
plt.show()

# 3. Test du Chi² : Vérification de l'indépendance entre Gravité et Type
contingency_table = pd.crosstab(df["Gravité"], df["Type"])
chi2, p_value, dof, expected = chi2_contingency(contingency_table)

print(f"📊 Test du Chi² :\n- Statistique Chi² : {chi2:.2f}\n- p-value : {p_value:.4f}")

# Interprétation du résultat
if p_value < 0.05:
    print("🔴 La relation entre Gravité et Type d’anomalie est significative (dépendante).")
    print("➡️ Il existe une corrélation entre la gravité d'une anomalie et son type.")
else:
    print("🟢 Pas de relation significative entre Gravité et Type d’anomalie.")
    print("➡️ La répartition des types d'anomalies est indépendante de leur gravité.")
```
